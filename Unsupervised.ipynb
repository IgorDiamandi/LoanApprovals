```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import warnings
from sklearn.cluster import AgglomerativeClustering
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("C:\\Users\\diana\\Downloads\\Loans\\LoanApprovals\\loans_clean4.csv", index_col=False)
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
X = df.loc[:,df.columns != 'Loan_Status']
```

```python
kmodel = KMeans(n_clusters=2, random_state=1).fit(X)
kmodel.labels_
```

```python
pd.crosstab(df['Loan_Status'],kmodel.labels_)
```

```python
hclmodel = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
hclmodel.fit_predict(X)
```

```python
pd.crosstab(df['Loan_Status'],hclmodel.labels_)
```

```python
sns.scatterplot(x=df['LoanAmount'], y=df['Credit_History'], hue=df['Loan_Status'])
```

```python
sns.scatterplot(x=df['LoanAmount'], y=df['Credit_History'], hue=hclmodel.labels_)
```

```python
sns.scatterplot(x=df['LoanAmount'], y=df['Credit_History'], hue=kmodel.labels_)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=2, random_state=1)
kmeans_labels = kmeans.fit_predict(X_scaled)

hcl = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
hcl_labels = hcl.fit_predict(X_scaled)

silhouette_kmeans = silhouette_score(X_scaled, kmeans_labels)
silhouette_hcl = silhouette_score(X_scaled, hcl_labels)

print(f"Silhouette Score for KMeans: {silhouette_kmeans}")
print(f"Silhouette Score for HCL: {silhouette_hcl}")
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', label='KMeans')
plt.title('KMeans Clustering with PCA')
plt.show()
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=hcl_labels, cmap='viridis', label='HCL')
plt.title('HCL Clustering with PCA')
plt.show()
```
